```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Start off by importing the cars.csv file in the jupyter notebook
cars = pd.read_csv("cars.csv")
```

```python
cars.head()
```

```python
#Generate a line plot graph for the column ‘model’ and ‘hp’.
#a. Map the ‘model’ column on the x-axis.
#b. Map the ‘hp’ column on the y-axis.
#c. Provide the x-axis label as Models of the cars.
#d. Provide the y-axis label as Horse-Power of Cars.
#e. Set the title as Model Names vs HorsePower.

x = cars["model"]
y = cars["hp"]

fig = plt.figure(figsize=(10,5))

plt.plot(x,y,linewidth=2.0,color='r',alpha=0.5,marker='o')
plt.title('Model Vs Hp')
plt.xlabel("Model")
plt.ylabel("hp")
plt.xticks(rotation=90)
plt.legend(["line"], loc="best")
plt.grid(True)
plt.show()
```

```python
#Assignment 2 

#Generate a bar plot graph for the columns ‘carbs’ and ‘gear’.
#a. Map the ‘carbs’ onto the x-axis.
#b. Map the ‘gear’ onto the y-axis.
#c. Provide the x-axis label as Number of carburetors.
#d. Provide the y-axis label as Number of forward gears.
#e. Set the title as carbs vs gear.

x = cars["carb"]
y = cars["gear"]

fig = plt.figure(figsize=(10,5))

plt.bar(x,y,linewidth=2.0,color='r')
plt.xlabel("carburetors")
plt.ylabel("forward gears")
plt.title("Carbs Vs Gears")
plt.show()
```

```python
#Assignment3 

#Plot a histogram for the column ‘wt’.
#a. Map the ‘wt’ onto the x-axis.
#b. Provide the x-axis label as ‘weight of the cars’.
#c. Provide the y-axis label as ‘Count’
#d. Set the number of bins as 30.
#e. Set the title as ‘Histogram for the weight values’

x = cars["wt"]

fig = plt.figure(figsize=(10,5))

plt.hist(x,bins=30)
plt.xlabel("weight of the carb")
plt.ylabel("Count")
plt.title("Histogram for the weight values")
plt.show()
```

```python
#Assignment 4 

#Plot a pie chart for columns: ‘cyl’ and ‘model’ form the mtcars.csv data frame.
%matplotlib inline
x = cars["cyl"]
y = cars["model"]

fig = plt.figure(figsize=(10,5))

plt.pie(x, labels=y, rotatelabels=True, radius=2)
plt.legend(loc="right", bbox_to_anchor=(3,0.5))
plt.show()
```

```python
#Assignment 5

#Plot the area chart for the colmns: ‘am’ and ‘carb’.
#a. Set the ‘am’ on the x-axis.
#b. Set the ‘carb’ on the y-axis.
#c. Provide the x-axis label as Transmission.
#d. Provide the y-axis labe as Number of carburetors.
#e. Provide the title as Transmission vs Number of carburetors. 

x = cars["am"]
y = cars["carb"]

fig = plt.figure(figsize=(10,5))

plt.fill_between(x,y)
plt.xlabel("Transmission")
plt.ylabel("carburetors")
plt.title("Transmission vs Number of carburetors")
plt.show()
```

```python
#Case Study Problem 1

customer = pd.read_csv("customer_churn.csv")
customer.head()
```

```python
#1. Sam has to build a bar-plot for the ‘Contract’ column
#a. Set the x-axis label to be ‘Contract Type of customer’
#b. Set the y-axis label to be ‘Count’
#c. Set the title of the plot to be ‘Distribution of Contract’
#d. Assign ‘orange’ color to all the bars

fig = plt.figure(figsize=(10,5))
x = customer["Contract"]

sns.countplot(x=x, data=customer)
# plt.bar(x,color="orange",linewidth=2.0)
plt.title("Distribution of Contract")
plt.xlabel("Contract Type of customer")
plt.ylabel("Count")
plt.show()
```

```python
#Sam has to build a histogram for the ‘MonthlyCharges’ column
#a. Set the x-axis label to be ‘Monthly Charges Incurred’
#b. Set the y-axis label to be ‘Count’
#c. Set the title of the plot to be ‘Distribution of Monthly Charges’
#d. Assign ‘forestgreen’ color to the bins

x = customer["MonthlyCharges"]

fig = plt.figure(figsize=(10,5))

plt.hist(x,bins=30,color="forestgreen")
plt.xlabel("Monthly Charges Incurred")
plt.ylabel("Count")
plt.title("Distribution of Monthly Charges")
plt.show()
```

```python
#Sam has to build a scatter-plot between ‘TotalCharges’ & ‘tenure’. ‘TotalCharges’ should be on
#the y-axis and ‘tenure’ should be on the x-axis
#a. Set the x-axis label to be ‘Tenure of the customer’
#b. Set the y-axis label to be ‘Total chargesIncurred’
#c. Set the title of the plot to be ‘Total Charges vs Tenure’
#d. Assign ‘indigo’ color to the points 

x = customer["tenure"].dropna().head(20)
y = customer["TotalCharges"].dropna().head(20)

fig = plt.figure(figsize=(10,5))

plt.scatter(x,y,color="indigo",linewidths = 2)
plt.xlabel("Tenure of the customer")
plt.ylabel("Total chargesIncurred")
plt.title("Total Charges vs Tenure")
plt.show()
```

```python
#Sam has to build a box-plot between ‘MonthlyCharges’ & ‘PaymentMethod’.
#‘MonthlyCharges’ should be on the y-axis and ‘PaymentMethod’ should be on the x-axis.
#a. Set the x-axis label to be ‘Payment Method of customer’
#b. Set the y-axis label to be ‘Monthly ChargesIncurred’
#c. Set the title of plot to be ‘Monthly Charges vs. Payment Method’
#d. Assign ‘olive’ color to the box-plots

import seaborn as sns

x = customer["PaymentMethod"]
y = customer["MonthlyCharges"]

fig = plt.figure(figsize=(10,5))
sns.boxplot(x = x,y = y, color='olive')
plt.xlabel("Payment Method of customer")
plt.ylabel("Monthly ChargesIncurred")
plt.title("Monthly Charges vs. Payment Method")
plt.show()
```

```python
#case study2

#Load cars data as dataframe using pandas and create a bar plot between number of cylinders
#and frequency of cars with that many number of cylinders.
#- Set xlabel as Number of cylinders.
#- Set ylabel as Frequency of cars.
#- Draw a bar plot.2.

x = cars["cyl"]

x.value_counts().plot(kind='bar')
plt.xlabel("Number of cylinders")
plt.ylabel("Frequency of cars")
plt.show()
```
